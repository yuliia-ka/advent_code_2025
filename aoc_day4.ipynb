```python
input_data = ['..@@.@@@@.','@@@.@.@.@@','@@@@@.@.@@','@.@@@@..@.','@@.@@@@.@@','.@@@@@@@.@','.@.@.@.@@@','@.@@@.@@@@','.@@@@@@@@.','@.@.@@@.@.']
```

```python
input_path = "C:\\Users\\ykazmin\\Desktop\\puzzle_input_day4.txt"

f = open(input_path).readlines()
input_data = [x.strip() for x in f]
input_data
```

```python
rows = len(input_data)
indexes = len(input_data[0])
input_data.insert(0,'.'*indexes)
input_data.append('.'*indexes)
for idx, elem in enumerate(input_data):
    #print('bulo',i)
    input_data[idx] = "." + elem + "."
    #print('stalo',i)
input_data = [list(a) for a in input_data]
```

```python
rolls = 0
for row in range(1,rows+1):
    for idx in range(1,indexes+1):
        if input_data[row][idx]=='@':
            cts = sum([input_data[row-1][idx-1] == '@',
                 input_data[row][idx-1] == '@',
                 input_data[row-1][idx] == '@',
                 input_data[row+1][idx+1] == '@',
                 input_data[row][idx+1] == '@',
                 input_data[row+1][idx] == '@',
                 input_data[row+1][idx-1] == '@',
                 input_data[row-1][idx+1] == '@'])

            if cts < 4:
                rolls += 1        
print(rolls)
```
