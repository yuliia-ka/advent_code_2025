```python
import numpy as np
import math

with open("C:\\Users\\ykazmin\\projects\\advent_code_2025\\puzzle_input_day8.txt", "r") as pup:
    input = pup.readlines()
    
input_lines = [a.replace('\n','') for a in input]

coordinates_tuples = [tuple(map(int,a.split(','))) for a in input_lines]
how_many = len(coordinates_tuples)

# Convert the list of tuples to a NumPy array
coordinates = np.array(coordinates_tuples, dtype=float)

# Calculate the distance matrix
def distance_matrix(coords):
    # Calculate the squared differences
    diff = coords[:, np.newaxis, :] - coords[np.newaxis, :, :]
    
    # Calculate the Euclidean distance
    dist_matrix = np.sqrt(np.sum(diff**2, axis=-1))
    #print(diff**2)
    return dist_matrix

# Get the distance matrix
distances = distance_matrix(coordinates)

# Extract the lower triangle indices
lower_triangle_indices = np.tril_indices(distances.shape[0], -1)

# Create a list of distances and their corresponding indices
distance_list = [(distances[i, j], (i, j)) for i, j in zip(lower_triangle_indices[0], lower_triangle_indices[1])]

# Sort the list by distance
sorted_distances = sorted(distance_list, key=lambda x: x[0])
```

```python
circuits = [set([a]) for a in range(how_many)]

for idx, (distance, indices) in enumerate(sorted_distances):
    if idx == 1000:
        break
    #print (idx,distance,indices) 
    
    elem1 = indices[0]
    elem2 = indices[1]
    # Remove sets that contain the specified value
    remove1 = [s for s in circuits if elem1 in s][0]
    remove2 = [s for s in circuits if elem2 in s][0]
    removed = remove1|remove2
    circuits =  [s for s in circuits if not any(value in s for value in {elem1,elem2})]
    circuits.append(removed)
    
circuit_lengths = [len(s) for s in circuits]
circuit_lengths.sort()

print("Part 1: ",math.prod((circuit_lengths[-3:])))
```
