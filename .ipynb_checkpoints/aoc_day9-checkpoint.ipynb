```python
import numpy as np
from itertools import combinations
import matplotlib.pyplot as plt

with open("C:\\Users\\ykazmin\\projects\\advent_code_2025\\puzzle_input_day9.txt", "r") as pup:
    input = pup.readlines()
    
input_lines = [a.replace('\n','') for a in input]

# 1. Convert input to a (N, 2) NumPy array
coords = np.array([list(map(int, line.split(','))) for line in input_lines])
coords = coords.astype(np.float64)

areas = []
# combinations(coords, 2) yields (point_a, point_b)
for p1, p2 in combinations(coords, 2):
    #find left coordinate
    x_axis = (p1[0],p2[0])
    y_axis = (p1[1],p2[1])
    left = min(x_axis)
    right = max(x_axis)
    bottom = max(y_axis)
    top = min(y_axis)
    area = (bottom- top + 1) * (right - left + 1)
    areas.append(area)
    
print("Part 1 : ",int(max(areas)))
```

```python
polygon_points = np.vstack([coords, coords[0]])

# Plotting to visualize
plt.figure(figsize=(8, 8))
plt.plot(polygon_points[:,0], polygon_points[:,1], 'r-o')
plt.fill(polygon_points[:,0], polygon_points[:,1], alpha=0.3)
plt.show()
polygon_points[:3]

# Create 'start' points and 'end' points
starts = polygon_points[:-1]
ends = polygon_points[1:]

# Stack them to get an array of shape (N, 2, 2)
# Each element is [[x_start, y_start], [x_end, y_end]]
poly_segments_np = np.stack((starts, ends), axis=1)
poly_segments_np[:3]

full_package = []
# combinations(coords, 2) yields (point_a, point_b)
for p1, p2 in combinations(coords, 2):
    #find left coordinate
    x_axis = (p1[0],p2[0])
    y_axis = (p1[1],p2[1])
    
    left = int(min(x_axis))
    right = int(max(x_axis))
    bottom = int(max(y_axis))
    top = int(min(y_axis))
    
    rect_coords = (
    [left, bottom],  # Corner 1
    [right, bottom], # Corner 2
    [right, top],    # Corner 3
    [left, top] )     # Corner 4

    area = int((bottom- top + 1) * (right - left + 1))
    full_package.append([area,rect_coords])
    
full_package.sort(reverse=True)
```

```python
def is_point_in_poly(x, y, poly):
    """Ray Casting algorithm. Returns True if point is inside or on the boundary."""
    n = len(poly)
    inside = False
    p1x, p1y = poly[0]
    for i in range(n + 1):
        p2x, p2y = poly[i % n]
        # Direct boundary check (important for integer grids)
        if y == p1y == p2y and min(p1x, p2x) <= x <= max(p1x, p2x): return True
        if x == p1x == p2x and min(p1y, p2y) <= y <= max(p1y, p2y): return True
        
        if (p1y > y) != (p2y > y):
            xints = (x - p1x) * (p2y - p1y) - (y - p1y) * (p2x - p1x)
            if (p2y > p1y): # Upward edge
                if xints < 0: inside = not inside
            else: # Downward edge
                if xints > 0: inside = not inside
        p1x, p1y = p2x, p2y
    return inside

def segments_intersect_exclusive(seg1, seg2):
    """
    Returns True ONLY if segments cross in the interior (X-shape).
    Shared endpoints or T-junctions are ignored.
    """
    (x1, y1), (x2, y2) = seg1
    (x3, y3), (x4, y4) = seg2

    def ccw(ax, ay, bx, by, cx, cy):
        val = (by - ay) * (cx - bx) - (bx - ax) * (cy - by)
        if abs(val) < 1e-9: return 0
        return 1 if val > 0 else -1

    o1 = ccw(x1, y1, x2, y2, x3, y3)
    o2 = ccw(x1, y1, x2, y2, x4, y4)
    o3 = ccw(x3, y3, x4, y4, x1, y1)
    o4 = ccw(x3, y3, x4, y4, x2, y2)

    # Standard intersection: points are on opposite sides
    if (o1 * o2 < 0 and o3 * o4 < 0):
        return True
    return False

# --- SEARCH ---
valid_rect_found = None

for area, rect_coords in full_package:
    # 1. Corner Check (Must be inside or on border)
    if not all(is_point_in_poly(c[0], c[1], polygon_points) for c in rect_coords):
        continue

    # 2. Prepare segments
    r_segs = [
        (rect_coords[0], rect_coords[1]), (rect_coords[1], rect_coords[2]),
        (rect_coords[2], rect_coords[3]), (rect_coords[3], rect_coords[0])
    ]

    # 3. Intersection Check (Strictly crossing)
    # We allow the rectangle to lay on top of polygon edges or share corners.
    has_crossing = False
    for r_seg in r_segs:
        for p_seg in poly_segments_np:
            if segments_intersect_exclusive(r_seg, p_seg):
                has_crossing = True
                break
        if has_crossing: break
    
    if not has_crossing:
        # Check one point in the middle to ensure we aren't in a 'hole'
        mid_x = (rect_coords[0][0] + rect_coords[2][0]) / 2
        mid_y = (rect_coords[0][1] + rect_coords[2][1]) / 2
        if is_point_in_poly(mid_x, mid_y, polygon_points):
            valid_rect_found = (area, rect_coords)
            break

print(f"Largest Valid Area: {valid_rect_found[0] if valid_rect_found else 'None'}")
```

```python
def get_intersections_vectorized(r_seg, poly_segments_np):
    """
    Uses vector math to check one rectangle segment against ALL polygon segments.
    Returns True if any 'X-crossing' is found.
    """
    # A, B are start and end of the rectangle segment
    A = np.array(r_seg[0])
    B = np.array(r_seg[1])
    
    # C is array of all poly starts, D is array of all poly ends
    C = poly_segments_np[:, 0]
    D = poly_segments_np[:, 1]

    def ccw_vec(P, Q, R):
        # Vectorized Cross Product
        val = (Q[..., 1] - P[..., 1]) * (R[..., 0] - Q[..., 0]) - \
              (Q[..., 0] - P[..., 0]) * (R[..., 1] - Q[..., 1])
        res = np.zeros(val.shape, dtype=int)
        res[val > 1e-9] = 1
        res[val < -1e-9] = -1
        return res

    # Orientations
    o1 = ccw_vec(A, B, C)
    o2 = ccw_vec(A, B, D)
    o3 = ccw_vec(C, D, A)
    o4 = ccw_vec(C, D, B)

    # Intersection occurs where orientations are opposite AND not zero (exclusive check)
    # This is a single NumPy boolean operation
    intersect_mask = (o1 * o2 < 0) & (o3 * o4 < 0)
    return np.any(intersect_mask)

# --- Optimized Loop ---
valid_rect_found = None

for area, rect_coords in full_package:
    # 1. Faster Bounding Box Check 
    # (If the rect's min/max x/y are outside the polygon's min/max, skip immediately)
    # [Already handled by your Point-in-Poly check]
    
    # 2. Corner Check
    if not all(is_point_in_poly(c[0], c[1], polygon_points) for c in rect_coords):
        continue

    # 3. Vectorized Segment Check
    r_segs = [
        (rect_coords[0], rect_coords[1]), (rect_coords[1], rect_coords[2]),
        (rect_coords[2], rect_coords[3]), (rect_coords[3], rect_coords[0])
    ]
    
    has_violation = False
    for rs in r_segs:
        if get_intersections_vectorized(rs, poly_segments_np):
            has_violation = True
            break
            
    if not has_violation:
        # Final Midpoint sanity check
        mid_x = (rect_coords[0][0] + rect_coords[2][0]) / 2
        mid_y = (rect_coords[0][1] + rect_coords[2][1]) / 2
        if is_point_in_poly(mid_x, mid_y, polygon_points):
            valid_rect_found = (area, rect_coords)
            break
            
valid_rect_found[0]
```

```python
# 2. Prepare Rectangle Data
# We add the first point to the end to close the rectangle loop
rect_raw = valid_rect_found[1]#((17416, 85761), (81313, 85761), (81313, 14720), (17416, 14720))
rect_points = np.vstack([rect_raw, rect_raw[0]])

# 3. Plotting
plt.figure(figsize=(10, 10))

# Plot the Big Polygon (Red)
plt.plot(polygon_points[:,0], polygon_points[:,1], 'r-o', markersize=2, label='Polygon Boundary')
plt.fill(polygon_points[:,0], polygon_points[:,1], color='red', alpha=0.2)

# Plot the Rectangle (Blue)
plt.plot(rect_points[:,0], rect_points[:,1], 'b-', linewidth=3, label='Test Rectangle')
plt.fill(rect_points[:,0], rect_points[:,1], color='blue', alpha=0.4)

# Formatting
plt.axis('equal') # Keeps the scale 1:1 so the shape isn't distorted
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title("Rectangle vs. Polygon Containment Check")

plt.show()
```
