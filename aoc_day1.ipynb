```python
input_data = ['L68','L30','R48','L5','R60','L55','L1','L99','R14','L82']
```

```python
input_path = "C:\\Users\\ykazmin\\Desktop\\puzzle_input_day1.txt"

f = open(input_path).readlines()
input_data = [x.strip() for x in f]
```

```python
#part 1
start = 50
pwd = 0
for string_i in input_data:
    direction = string_i[0]
    counts = int(string_i[1:])
    #print(direction, counts)
    if direction == 'R':
        start += counts
    
    else:
        start -= counts
        
    remainder = start % 100 
    if remainder == 0:
        pwd += 1  
        
pwd
```

```python
def move_the_clock(start,direction,how_much):
    if direction == 'R':
        start += how_much
    else:
        start -= how_much
    return start

def count_zeroliashas(whatever_the_fuck):
    if whatever_the_fuck > 0:
        return whatever_the_fuck // 100
    else:
        return abs(whatever_the_fuck) // 100
```

```python
start = 50
pwd = 0

for string_i in input_data:
    #print('Start at: ', start)
    new_value = move_the_clock(start,string_i[0],int(string_i[1:]))
    pwd += count_zeroliashas(new_value)
    start = new_value % 100

pwd
```

```python
#part 2
location = 50
pwd = 0

for string_i in input_data:
    
    direction = string_i[0]
    counts = int(string_i[1:])
    init_loc = location
   
    if direction == 'R':
        location += counts
    else:
        location -= counts
    
    if location > 0:
        factor = location // 100
    else:
        factor = abs(location) // 100
        if init_loc != 0:
            factor += 1
            
    pwd += factor
    
    location = location % 100
    
    print(direction,counts)
    print('Crossed zero times: ', factor)
    print("New starting point: ",location)
    print("# accumulated zeros: ", pwd)
    print("-----------")
    
pwd
```
