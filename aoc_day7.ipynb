```python
with open("C:\\Users\\ykazmin\\projects\\advent_code_2025\\puzzle_input_day7.txt", "r") as pup:
    input = pup.readlines()
    
input_lines = [a.replace('\n','') for a in input]

where_s = input_lines[0].index('S')
beams_list = [Counter([(0,where_s)])]

splitting_points = [
    (row, col)
    for row, line in enumerate(input_lines)
    for col, char in enumerate(line)
    if char == "^"
]

beams_list
```

```python
from collections import Counter

for line in range(len(input_lines)-1):
    
    curr_beams = beams_list[line]
    
    future_beams = Counter()
    
    for beam,counts in curr_beams.items():
    
        curr_pos = beam  
        next_pos = (curr_pos[0]+1,curr_pos[1])
        content = input_lines[next_pos[0]][next_pos[1]]

        if content == '.':
            future_beams[next_pos] += counts

        else:
            left = (next_pos[0],next_pos[1]-1)
            right = (next_pos[0],next_pos[1]+1)
            future_beams[left] += counts
            future_beams[right] += counts
            
    beams_list.append(future_beams)
    
print("Answer to part 2: ",sum(beams_list[-1].values()))
```
