```python
input_data = ['987654321111111','811111111111119','234234234234278','818181911112111']
```

```python
input_path = "C:\\Users\\ykazmin\\Desktop\\puzzle_input_day3.txt"

f = open(input_path).readlines()
input_data = [x.strip() for x in f]
input_data
```

```python
#part 1

keys = []

for string_i in input_data:
    split_str_list = list(string_i)
    length_list = len(split_str_list)
    
    mapped_int = [int(a) for a in split_str_list]
    
    first_digit = max(mapped_int)
    loc_max = mapped_int.index(first_digit)
    if (loc_max+1==length_list):
        first_digit = max(mapped_int[:-1])
            
    where_largest = mapped_int.index(first_digit)
    second_largest = max(mapped_int[(where_largest+1):])
    
    ready_number = first_digit*10+second_largest
    keys.append(ready_number)
    
sum(keys)
```

```python
#part 2

keys = []

for string_i in input_data:
    
    split_str_list = list(string_i)
    mapped_int = [int(a) for a in split_str_list]
    
    digits = []
    
    for digit in range(1,13):
        if digit == 12:
            found_digit = max(mapped_int)
        else:
            found_digit = max(mapped_int[:-(12-digit)])
        #print(found_digit)
        where_found_digit = mapped_int.index(found_digit)
        mapped_int = mapped_int[where_found_digit+1:]
        #print(mapped_int)
        digits.append(found_digit)
    #print(int("".join(map(str, digits))))
    ready_key = int("".join(map(str, digits)))
    keys.append(ready_key)
        
sum(keys)
```
