```python
input_path = "C:\\Users\\ykazmin\\Desktop\\puzzle_input_day2.txt"

f = open(input_path).readlines()
input_data = f[0].split(",")
input_data
```

```python
#part 1

overlaps = []

for string_i in input_data:
    
    lower_bound = int(string_i.split('-')[0])
    upper_bound = int(string_i.split('-')[1])
    
    given_list = list(range(lower_bound,upper_bound+1))
    #print(given_list)
    
    len_lower_bound = len(str(lower_bound))
    len_upper_bound = len(str(upper_bound))
    
    start = min(len_lower_bound,len_upper_bound)
    end = max(len_lower_bound,len_upper_bound)
    
    if start % 2 != 0:
        start += 1
        
    if end % 2 != 0:
        end -= 1
        
    start_pair_len = start / 2
    end_pair_len = end / 2
    start_pair_len = int(start_pair_len)
    end_pair_len = int(end_pair_len)
    #print(start_pair_len,end_pair_len)
    
    search_list = []
    
    for curr_length in list(range(start_pair_len,end_pair_len+1)):
    
        # The smallest d-digit number is 10^(d-1)
        smallest_number = 10**(curr_length - 1)
        if smallest_number == 0:
            smallest_number +=1

        # The largest d-digit number is 10^d - 1
        largest_number = 10**curr_length - 1
        
        half_list = list(range(smallest_number,largest_number+1))
        duplicated_list = [int(str(a)*2) for a in half_list]
        
        search_list.extend(duplicated_list)
        
    overlaps.extend(list(set(given_list) & set(search_list)))    
        
print(sum(overlaps))
```
